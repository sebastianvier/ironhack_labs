```python
## import the data
path = '../lab-data-cleaning-and-wrangling/'
X_train = pd.read_csv(path + 'X_train.csv')
X_test  = pd.read_csv(path + 'X_test.csv')
y_train = pd.read_csv(path + 'y_train.csv')
y_test  = pd.read_csv(path + 'y_test.csv')
```

```python
print(X_train.head())
print(X_test.head())
print(y_train.head())
print(y_test.head())
```

```python
X_test
```

```python
Ln = LinearRegression().fit(X_train, y_train)
score = Ln.score(X_test, y_test)
score 
```

```python
kn_model = KNeighborsRegressor(n_neighbors=3).fit(X_train, y_train)
score = kn_model.score(X_test, y_test)
score 
```

```python
for i in range(1,12):
    kn_model = KNeighborsRegressor(n_neighbors=i).fit(X_train, y_train)
    score = kn_model.score(X_test, y_test)
    print(f'Score for n_neighbors {i}: {score}')
```

```python
kn_model = KNeighborsRegressor(n_neighbors=5, weights='distance', p=1).fit(X_train, y_train)
score = kn_model.score(X_test, y_test)
score 
```

```python
for i in range(1,12):
    kn_model = KNeighborsRegressor(n_neighbors=i, p=1, weights='distance').fit(X_train, y_train)
    score = kn_model.score(X_test, y_test)
    print(f'Score for n_neighbors {i}: {score}')
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
for n in [1, 5, 10]:
    regr = MLPRegressor(random_state=1, max_iter=n).fit(X_train, y_train)
    regr.score(X_test, y_test)
    print(f'Score for max_iter {n}: {score}')
```
