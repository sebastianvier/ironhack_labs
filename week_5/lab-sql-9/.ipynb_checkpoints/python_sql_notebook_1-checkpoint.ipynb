#### Common used functions

```python
def get_table_fields(table_name):
    return [i[0] for i in list(engine.execute("show fields from " + table_name + ";"))]
    # The space after the from is crucial for this to work.
```

```python
statement = "SELECT * FROM rental "
statement += "where MONTH(RENTAL_DATE) = 5;"
pd.DataFrame(engine.execute(statement))
```

Exercise #3
- Check the number of rentals for each customer for May.
- Check the number of rentals for each customer for June.

```python
statement = "SELECT * FROM rental_may;"
may_rentals = pd.DataFrame(engine.execute(statement), columns=get_table_fields("rental_may"))
statement = "SELECT * FROM rental_june;"
june_rentals = pd.DataFrame(engine.execute(statement), columns=get_table_fields("rental_june"))
```

```python
may_rentals.head()
```

```python
# May rentals
may_rentals.groupby('customer_id').count()[['rental_id']].sort_values('rental_id', 
                                                                       ascending=False)
```

```python
# June rentals
june_rentals.groupby('customer_id').count()[['rental_id']].sort_values('rental_id', 
                                                                       ascending=False)
```
