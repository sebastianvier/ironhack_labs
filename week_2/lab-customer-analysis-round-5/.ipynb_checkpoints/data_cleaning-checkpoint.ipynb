```python
import time
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("../../data_files/csv_files/marketing_customer_analysis.csv")
data.columns = [i.lower().replace(" ", "_") for i in data.columns]
```

```python
data_numeric = data.select_dtypes(include=np.number)
data_categorical = data.select_dtypes(include="object")
```

```python
def axes_generator(x,y):
    lst1 = [i for i in range(x)]
    lst2 = [i for i in range(y)]
    for fnum in lst1:
        for snum in lst2:
            yield fnum, snum
```

```python
ordered_pairs = [i for i in axes_generator(4,2)]

fig, axes = plt.subplots(4,2, figsize=(10,25))

for i in range(len(data_numeric.columns)):
    axes[ordered_pairs[i]].hist(data[data_numeric.columns[i]], bins=40)
    axes[ordered_pairs[i]].set_title(data_numeric.columns[i])

plt.show()
```

```python
def transformation(col):
    return np.log(col +1)
```

```python
ordered_pairs = [i for i in axes_generator(4,2)]

fig, axes = plt.subplots(4,2, figsize=(10,25))

for i in range(len(data_numeric.columns)):
    column = data_numeric[data_numeric.columns[i]].map(transformation)
    axes[ordered_pairs[i]].hist(column, bins=80)
    axes[ordered_pairs[i]].set_title(data_numeric.columns[i])

plt.show()
```

```python
data_numeric = data_numeric.applymap(transformation)
```

I want to normalize: customer_lifetime_value, total_claim_amount, monthly_premium_auto, income

```python
scaler = preprocessing.StandardScaler().fit(data_numeric)
scaled_numeric = scaler.transform(data_numeric)
scaled_numeric = pd.DataFrame(scaled_numeric)
scaled_numeric.columns = data_numeric.columns
```

```python
ordered_pairs = [i for i in axes_generator(4,2)]

fig, axes = plt.subplots(4,2, figsize=(10,25))

for i in range(len(scaled_numeric.columns)):
    axes[ordered_pairs[i]].hist(scaled_numeric[scaled_numeric.columns[i]], bins=50)
    axes[ordered_pairs[i]].set_title(scaled_numeric.columns[i])

plt.show()
```

```python
scaled_numeric = scaled_numeric.drop(["months_since_last_claim",
                     "months_since_policy_inception",
                     "months_since_policy_inception",
                     "number_of_open_complaints",
                     "number_of_policies"],axis=1)
```

```python
ordered_pairs = [i for i in axes_generator(4,2)]

fig, axes = plt.subplots(2,2, figsize=(10,10))

for i in range(len(scaled_numeric.columns)):
    axes[ordered_pairs[i]].hist(scaled_numeric[scaled_numeric.columns[i]], bins=50)
    axes[ordered_pairs[i]].set_title(scaled_numeric.columns[i])

plt.show()
```

```python
y = scaled_numeric["total_claim_amount"] 
X = scaled_numeric.drop(["total_claim_amount"] , axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
regression = LinearRegression().fit(X_train, y_train)
print(regression.score(X,y))
prediction = regression.predict(X_test)
```

```python
mean_squared_error(y_test, prediction)
```
