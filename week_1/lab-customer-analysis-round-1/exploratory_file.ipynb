```python
def customer_lifetime_to_numeric(df):
    def change_to_float(x):
        if type(x) == str:
            return float(x[:-1])/100
        else:
            return x
           
    return df.customer_lifetime_value.map(lambda x:change_to_float(x))

data.customer_lifetime_value = customer_lifetime_to_numeric(data)
```
