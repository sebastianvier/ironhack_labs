##### 5. Build the logistic regression model.

```python
## Extract variables and split the data
y = data.churn
X = data['tenure senior_citizen monthly_charges'.split(' ')]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=39)
```

```python
## We are going to fit the scaler on the train data because then 
## the data on the test is not affecting the scaler
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
```

```python
predictions = clf.predict(X_test)
```
